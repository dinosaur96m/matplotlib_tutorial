```python
import pylab as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
plt.plot(t, s, linestyle='--')

# plt.xticks([0,.25,2,3,4,6,8])
plt.yticks([0,1,2])

plt.show()
```
