```python
import pylab as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
plt.plot(t, s, linestyle='--')

plt.show()
```
