### Basic Graph

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

xs = [0,1,2,3,4]
ys = [4,5,6,7,8]

x2 = [0, 2, 4, 6, 8]
y2 = [0, 4, 8, 16, 32]

plt.plot(xs, ys, label='2x', color='red', linewidth=5, marker='.', linestyle='--', markersize='10')
plt.plot(x2, y2, 'b^--', label='2^2' )

plt.title("Our First Graph!", fontdict={'fontname': 'Comic Sans MS', 'color': 'white' })
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4,6,8])
plt.yticks([0, 2, 4, 6,8, 10, 20, 30])

plt.legend()

plt.show()
```
