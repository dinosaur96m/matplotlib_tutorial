### Basic Graph

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# dpi is pixels per inch
plt.figure(figsize=(4,1.5), dpi=300)

xs = [0,1,2,3,4]
ys = [4,5,6,7,8]

x2 = [0, 2, 4, 6, 8]
y2 = [0, 4, 8, 16, 32]

# line 3
x3 = np.arange(0, 4.5, 0.5)
print(x3)
plt.plot(x3, x3**2, 'r', label='X^2')


plt.plot(xs, ys, label='2x', color='red', linewidth=5, marker='.', linestyle='--', markersize='10')
plt.plot(x2, y2, 'b^--', label='2^2' )

plt.title("Our First Graph!", fontdict={'fontname': 'Comic Sans MS', 'color': 'white' })
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4,6,8])
plt.yticks([0, 2, 4, 6,8, 10, 20, 30])

plt.legend()

# save your graph
# plt.savefig('mygraph.png', dpi=300)

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```
